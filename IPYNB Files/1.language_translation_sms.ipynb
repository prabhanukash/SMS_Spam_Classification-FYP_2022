```python
import pandas as pd

data = pd.read_excel('Final_Dataset_excel_language.xlsx')
data.head(-1)
```

```python
from lingua import Language, LanguageDetectorBuilder
languages = [Language.ENGLISH, Language.HINDI, Language.TELUGU]


language_labels = []
for index, row in data.iterrows():
  text = row['message']
  print(text)
  words = len(text.split(' '))
  if words > 1:
    try:
      detector = LanguageDetectorBuilder.from_languages(*languages).with_minimum_relative_distance(0.25).build()
      label = detector.detect_language_of(text)
      print(index)
      print(label)
      label = str(label).split('.')[1].lower()
      language_labels.append(label)
    except Exception as e:
      language_labels.append('n/a')
  else:
    language_labels.append('n/a')

data['language'] = language_labels

data.head(-1)
```

```python
translations = []
for index, row in data.iterrows():
  text = row['message']
  lang = row['language']
  if lang == 'telugu' or lang == 'hindi':
    res = translator.translate(text)
    trans = res.text
    translations.append(trans)
  else:
    translations.append(text)

data['translated_text'] = translations

data.head(-1)


```
